```python
''' Proper no. of Dimensions for the dataset to be reduced into - to reduce Curse of Dim. and reduce Noise in Data and preserve maximum variance ''' 

from sklearn.decomposition import PCA


dim = []
sum= 0

for n in range(1,20) :
    pca = PCA(n_components = n)
    pca.fit(X1)
    
    t = list(pca.explained_variance_ratio_) # y-value - proportion of total variance preserved
    sum+=t[-1]
    dim.append(sum)

plt.plot(range(1,20) , dim , marker='p', linestyle='-', color='b')
plt.xlabel("No of dim to reduce into")
plt.ylabel("Cumm. Explained Variance Ratio")
plt.show()

print("CONCLUSION - ELBOW POPINT AT N = 10 , HENCE Optimum No. of Dim  = 10 for dataset to be reduced into ")
```

### Data Visualization

### K-Means clustering


```python
'''to find appropriate no of clusters in the dataset X (of 10 dim) '''

# use model's WCSS - WITHIN CLUSTER SUM OF SQUARES .

from sklearn.cluster import KMeans
wcss =[]

# Since , max 33 patients individual cluster

for k in range(1 ,33) :
    kmeans = KMeans(n_clusters = k , init = "k-means++")
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

#plot the wcss v/s no of clusters graph 
plt.plot(np.arange(1 , 33) , wcss , marker='o' , linestyle ='-' , color ='b')
plt.xlabel("No of clusters- k for 10 DIM. ")
plt.ylabel("WCSS")
plt.show()
# wcss =sum of squared distance of daat ponts from its clustr's mean , and sum this over all clusters. 
# lesser the wcss , bettter the cluster 

# elbow joint shows where model inertia / wcss start to decrement at slower rate

print("CONCLUSION - ELBOW POPINT AT K = 10 , HENCE by - KMeans Clustering , Optimum No. of Clusters = 10")
```

### Heirarchical Clustering - Agglomerative Clustering 

```python
''' 
In heirarchichal clustering - all individual data points are trated initially as own  distict clusters ,and sequentially , we go on
 adding the nearest clusters at each step , till all clusters merge into 1 single biggest cluster . 

 stop ats tep when appropriate no of clusters (acc. to you as programmer) are formed . 
e.g. - agglomerative clustering .
'''

from sklearn.cluster import AgglomerativeClustering 

# FInding  optimum clusters for Agg clustering 

from sklearn.metrics import silhouette_score  as score

lis = []

for k in range(2, 33) :
    model = AgglomerativeClustering(n_clusters=k)
    
    res = model.fit_predict(X)
    lis.append(score(X , res))


plt.plot(range(2 , 33) , lis , marker ='o' , linestyle = '-' , color='g')
plt.ylabel("silhouette score")
plt.xlabel("No. of Clusters")
plt.axvline(x = 9 ,color = "r" )
print("Optimum NO. of CLusters suggesated by - Agglomerative Clustering = 9 clusters .")

plt.plot()


```

```python
dendogram = AgglomerativeClustering(n_clusters = 9)
pred_heir = dendogram.fit_predict(X)
```
